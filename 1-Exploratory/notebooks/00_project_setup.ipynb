## SEÇÃO 3 — NLP Methodology

### 3.1 BM25 — Source Relevance Ranking

**BM25** (Best Matching 25) is the primary ranking algorithm used to score each of the 20 sources by FII content relevance.

**Why BM25 over raw TF-IDF for ranking**:
- **Term saturation**: high-frequency terms plateau (prevents spam dominance)
- **Document length normalization**: avoids unfairly rewarding longer articles
- **Query-specific**: scores are computed against FII-specific queries (e.g., `P/VP dividendo vacância yield`)
- **XAI-aligned**: scores decompose per term — fully interpretable and auditable

**Role in this project**: `BM25 = ranking layer`. TF-IDF = vectorization layer for topic modeling and negative context.

*Full mathematical derivation, hyperparameter analysis, and Robertson & Zaragoza (2009) reference:*
→ `docs/BM25_FOUNDATION.md`

### 3.2 Sentiment Analysis Architecture (PT-BR)

TextBlob alone is **insufficient** for Brazilian Portuguese financial text. NB03 implements a 4-layer approach:

| Layer | Method | Purpose |
|---|---|---|
| **Layer 1** | TextBlob baseline | Initial polarity signal |
| **Layer 2** | PT-BR financial lexicon | Domain-specific term weighting |
| **Layer 3** | Rule-based contextual correction | Negation, intensifiers, FII idioms |
| **Layer 4** | Negative Context Detection | ★ Desafio Extra — co-occurrence analysis |

**Limitation**: Combined accuracy ~75–80% on PT-BR financial text. Documented in NB05 governance section.

````python
print("\n⚡ PySpark Initialization")

spark = None
spark_available = False
SPARK_DRIVER_MEMORY = os.getenv('SPARK_DRIVER_MEMORY', '4g')

try:
    import sys, os
    os.environ['PYSPARK_PYTHON'] = sys.executable
    os.environ['PYSPARK_DRIVER_PYTHON'] = sys.executable
    from pyspark.sql import SparkSession
    from pyspark import SparkConf
    from pyspark.sql.functions import udf, col
    from pyspark.sql.types import StringType

    existing = SparkSession.getActiveSession()
    if existing:
        existing.stop()
        print('   ℹ️  Stopped existing Spark session')

    conf = SparkConf() \
        .setAppName('Investor-Intelligence-Platform-FIIs') \
        .setMaster('local[*]') \
        .set('spark.driver.memory', SPARK_DRIVER_MEMORY) \
        .set('spark.sql.shuffle.partitions', '8') \
        .set('spark.sql.adaptive.enabled', 'false') \
        .set('spark.ui.showConsoleProgress', 'false')

    spark = SparkSession.builder.config(conf=conf).getOrCreate()
    spark.sparkContext.setLogLevel('WARN')

    # Verify config not silently inherited
    assert spark.conf.get('spark.sql.shuffle.partitions') == '8'
    assert spark.conf.get('spark.sql.adaptive.enabled') == 'false'

    # ── Test 1: DataFrame + aggregation (NB02–NB05) ─────────────────────
    _df = spark.createDataFrame(
        [('FII', 1, 'infomoney.com.br'), ('dividendo', 2, 'fundsexplorer.com.br')],
        ['term', 'count', 'source']
    )
    _total = _df.agg({'count': 'sum'}).collect()[0][0]
    assert _total == 3
    print('   ✅ DataFrame + aggregation')

    # ── Test 2: UDF registration (NB02 HTML cleaning) ───────────────────
    _udf = udf(lambda x: x.upper() if x else '', StringType())
    _result = _df.withColumn('term_upper', _udf(col('term'))).collect()
    assert _result[0]['term_upper'] == 'FII'
    print('   ✅ UDF registration + execution')

    # ── Test 3: RDD parallelize / flatMap / reduceByKey (NB01 word count) ─
    _rdd = spark.sparkContext.parallelize(
        ['fii dividendo yield', 'fii vacância gestão']
    )
    _wc = _rdd.flatMap(lambda x: x.split()) \
              .map(lambda w: (w, 1)) \
              .reduceByKey(lambda a, b: a + b) \
              .collect()
    _wc_dict = dict(_wc)
    assert _wc_dict.get('fii') == 2, f'Word count error: {_wc_dict}'
    print('   ✅ RDD parallelize / flatMap / map / reduceByKey')

    # ── Test 4: Parquet write + read (NB02–NB05) ─────────────────────────
    import tempfile, shutil
    _tmpdir = PROJECT_ROOT / 'data' / 'bronze' / '_spark_test'
    try:
        _df.write.mode('overwrite').parquet(str(_tmpdir))
        _df_read = spark.read.parquet(str(_tmpdir))
        assert _df_read.count() == 2
        print('   ✅ Parquet write + read')
    finally:
        shutil.rmtree(str(_tmpdir), ignore_errors=True)

    spark_available = True
    print(f'\n   ✅ Spark {spark.version} | All critical operations validated')
    print(f'      memory={SPARK_DRIVER_MEMORY} | partitions=8 | adaptive=false')

except Exception as e:
    err_type = type(e).__name__
    print(f'   ⚠️  Spark [{err_type}]: {str(e)[:120]}')
    print('      Spark required for NB01–NB05. Install JDK 11+ and pyspark.')
````

````python
# ─── Generate dashboard/chatbot/groq_client.py (disclaimer enforcement) ────
(PROJECT_ROOT / 'dashboard' / 'chatbot').mkdir(parents=True, exist_ok=True)
(PROJECT_ROOT / 'dashboard' / '__init__.py').touch()
(PROJECT_ROOT / 'dashboard' / 'chatbot' / '__init__.py').touch()

(PROJECT_ROOT / 'dashboard' / 'chatbot' / 'groq_client.py').write_text('''
"""
dashboard/chatbot/groq_client.py
Groq AI client with mandatory financial disclaimer enforcement.
Humanistic AI: educational and analytical only. Never financial advisory.
"""

import os

DISCLAIMER = (
    "\\n---\\n"
    "⚠️ **AVISO IMPORTANTE**: Esta resposta possui caráter exclusivamente "
    "**educacional e analítico** e **não constitui recomendação de investimento**. "
    "Consulte um assessor financeiro qualificado antes de tomar decisões."
)

FORBIDDEN_ADVISORY_TERMS = [
    "recomendo comprar", "recomendo vender", "compre agora",
    "venda agora", "melhor investimento", "garanto retorno",
    "invista em", "compra de fii", "venda de fii",
]


def get_api_key() -> str:
    """Load GROQ_API_KEY_FII from st.secrets or environment. Never hardcoded."""
    try:
        import streamlit as st
        return st.secrets["GROQ_API_KEY_FII"]
    except Exception:
        key = os.getenv("GROQ_API_KEY_FII", "")
        if not key:
            raise EnvironmentError(
                "GROQ_API_KEY_FII not configured. "
                "Add to .streamlit/secrets.toml or set as environment variable."
            )
        return key


def contains_advisory_language(text: str) -> bool:
    """Detect if model response contains forbidden advisory language."""
    t = text.lower()
    return any(term in t for term in FORBIDDEN_ADVISORY_TERMS)


def chat(user_message: str, context: str, model: str = "llama-3.1-8b-instant") -> str:
    """
    Send message to Groq. Appends DISCLAIMER to every response.
    Raises ValueError if response contains forbidden advisory language.
    """
    from groq import Groq
    client = Groq(api_key=get_api_key())

    system_prompt = (
        "Você é um assistente analítico especializado em Fundos de Investimento Imobiliário (FII) "
        "para fins EXCLUSIVAMENTE educacionais e analíticos.\\n\\n"
        "REGRAS ABSOLUTAS:\\n"
        "- NUNCA recomende compra ou venda de FIIs\\n"
        "- NUNCA forneça consultoria financeira\\n"
        "- SEMPRE mantenha caráter educacional\\n"
        "- Use apenas dados do contexto fornecido\\n\\n"
        f"CONTEXTO DOS DADOS:\\n{context}"
    )

    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user",   "content": user_message},
        ],
        max_tokens=500,
        temperature=0.7,
    )

    answer = response.choices[0].message.content

    if contains_advisory_language(answer):
        answer = (
            "[Resposta filtrada: conteúdo de consultoria detectado. "
            "Esta plataforma não oferece recomendações de investimento.]"
        )

    return answer + DISCLAIMER  # Disclaimer ALWAYS appended
''')
# (verified below)

# Verify generated file is importable (syntax check)
_gc_path = PROJECT_ROOT / 'dashboard' / 'chatbot' / 'groq_client.py'
import ast as _ast
_ast.parse(_gc_path.read_text())
print('✅ dashboard/chatbot/groq_client.py (syntax verified, disclaimer enforced)')
````
